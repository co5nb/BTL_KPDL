```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
palette = {'color': ['#440154', '#bc3754', '#21918c', '#a0da39', '#fde725'],
           'Clusters': [0, 1, 2, 3, 4]}

palette['color']
```

```python
data = pd.read_csv("../completed_data.csv")
data_2 = pd.read_csv("../completed_data_noScaling.csv")
# Giá trị tối thiểu và tối đa ban đầu của từng cột
min_max_values = {
    'age': (data_2['age'].min(), data_2['age'].max()),
    'happiness.score': (data_2['happiness.score'].min(), data_2['happiness.score'].max()),
    'sex': (data_2['sex'].min(), data_2['sex'].max()),
    'depression_severity': (data_2['depression_severity'].min(), data_2['depression_severity'].max())
}
columns_to_scale = ['age', 'happiness.score', 'sex', 'depression_severity']
# Chuyển ngược dữ liệu chuẩn hóa về giá trị ban đầu
for column in columns_to_scale:
    min_val, max_val = min_max_values[column]
    data[column] = data[column] * (max_val - min_val) + min_val

```

```python
plt.figure(figsize = (9,4))

sns.countplot(x=data['cluster'], edgecolor = 'k', palette = palette['color'])

plt.tight_layout()
plt.show()
```

```python
data_fem = data[data['sex'] == 0]
data_mal = data[data['sex'] == 1]
```

```python
plt.figure(figsize = (15,6))

ax = plt.subplot(1,2,1)
sns.scatterplot(data = data_fem, x = 'depression_severity', y = 'age', hue = 'cluster',
                palette = ['#21918c', '#fde725'],
                ax = ax, s = 100)
plt.title('Women')

ax = plt.subplot(1,2,2)
sns.scatterplot(data = data_mal, x = 'depression_severity', y = 'age', hue = 'cluster',
                palette = ['#440154', '#bc3754','#a0da39'],
                ax = ax, s = 100)
plt.title('Men')

plt.tight_layout()
plt.show()
```

```python
sns.set_context(context = 'paper', font_scale= 1)
plt.figure(figsize = (11,9))

ax = plt.subplot(4,2,1)
sns.histplot(x = 'age', data = data_fem, hue = 'cluster', 
             palette = ['#21918c', '#fde725'], 
             bins = 10, alpha = 0.8, edgecolor = 'k', ax = ax)

plt.title('Women')

ax = plt.subplot(4,2,2)
sns.histplot(x = 'age', data = data_mal, hue = 'cluster', 
             palette = ['#440154', '#bc3754','#a0da39'], 
             bins = 10, alpha = 0.8, edgecolor = 'k', ax = ax)

plt.title('Men')

ax = plt.subplot(4,2,3)
sns.histplot(x = 'happiness.score', data = data_fem, hue = 'cluster', 
             palette = ['#21918c', '#fde725'], 
             bins = 10, alpha = 0.8, edgecolor = 'k', ax = ax,
            legend = False)

ax = plt.subplot(4,2,4)
sns.histplot(x = 'happiness.score', data = data_mal, hue = 'cluster', 
             palette = ['#440154', '#bc3754','#a0da39'], 
             bins = 10, alpha = 0.8, edgecolor = 'k', ax = ax,
            legend = False)

ax = plt.subplot(4,2,5)
sns.histplot(x = 'depression_severity', data = data_fem, hue = 'cluster', 
            palette = ['#21918c', '#fde725'], 
             bins = 10, alpha = 0.8, edgecolor = 'k', ax = ax,
            legend = False)

ax = plt.subplot(4,2,6)
sns.histplot(x = 'depression_severity', data = data_mal, hue = 'cluster', 
             palette = ['#440154', '#bc3754','#a0da39'], 
             bins = 10, alpha = 0.8, edgecolor = 'k', ax = ax,
            legend = False)
plt.tight_layout()
plt.show()
```
